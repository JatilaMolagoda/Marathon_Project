```python
#cleaning the data
```

```python
#Lets analys the races in Italy, 2020 and 50km distance
```

```python
df[df['Event distance/length'] == '50km']
```

```python
#to get the races held in Italy must extract the country code
```

```python
df['Country Code'] = df['Event name'].str.extract(r'\((\w{3})\)')
```

```python
df.head(5)
```

```python
#filtering our datset
```

```python
df[(df['Country Code'] == 'ITA')  & (df['Event distance/length'] == '50km') & (df['Year of event'] == 2022)]
```

```python
#creating a new data frame to get only the data we need
```

```python
df2 = df[(df['Country Code'] == 'ITA')  & (df['Event distance/length'] == '50km') & (df['Year of event'] == 2022)]
```

```python
df2.head()
```

```python
df2.shape
```

```python
#remove the counrty code from the event name so it will easy to read
```

```python
df2['Event name'] = df2['Event name'].str.replace(r'\s\(\w{3}\)', '', regex=True)
```

```python
df2.tail()
```

```python
#creating athlete age column
```

```python
df2['Athlete age'] = 2022 - df2['Athlete year of birth']
```

```python
df2.head()
```

```python
#clearing the h from Athlete performance
```

```python
df2['Athlete performance'] = df2['Athlete performance'].str.replace(' h', '', regex=False)
```

```python
df2['Athlete performance']
```

```python
df2.head(1)
```

```python
#drop some columns to make our data set mor nicer
```

```python
df2 = df2.drop(['Athlete year of birth', 'Athlete age category', 'Country Code', 'Athlete club', 'Year of event' ], axis=1, errors='ignore')
```

```python
df2.head(2)
```

```python
#clean up null values
```

```python
df2.isna().sum()
```

```python
#checking the details that have null values
```

```python
df2[df2['Athlete age'].isna() == 1]
```

```python
#lets drop the null values
```

```python
df2 = df2.dropna()
```

```python
df2.isna().sum()
```

```python
#lets check for duplicates
```

```python
df2[df2.duplicated() == True]
```

```python
#there is no dupes, so lets reset index
```

```python
df2.reset_index(drop = True)
```

```python
#fixing the data types
```

```python
df2.dtypes
```

```python
df2['Athlete age'] = df2['Athlete age'].astype(int)
```

```python
df2['Athlete average speed'] = df2['Athlete average speed'].astype(float)
```

```python
df2.dtypes
```

```python
#finding insights from the data
```

```python
#Q1. mean average speed of male and female
```

```python
df2.groupby(['Athlete gender'])['Athlete average speed'].mean()
```

```python
#Q2. What age groups are the best in 50km races(20+ races min)
```

```python
df2.groupby('Athlete age')['Athlete average speed'].agg(['mean', 'count']).sort_values('mean', ascending = False).query('count>19')
```

```python
#Q3.Which season is the best to run?
```

```python
#lets extract the month from event date
```

```python
df2['race month'] = df2['Event dates'].str.split('.').str.get(1).astype(int)
```

```python
df2['race season'] = df2['race month'].apply(lambda x: 'winter' if x > 11 else 'fall' if x > 8 else 'Summer' if x > 5 else 'Spring' if x > 2 else 'winter')
```

```python
df2.groupby('race season')['Athlete average speed'].agg('mean', 'count')
```
